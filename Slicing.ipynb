# **Slicing**

Slicing is the extraction of a part of a string, list, or tuple. It enables users to access the specific range of elements by mentioning their indices.

Syntax: Object [start:stop:step]

“Start” specifies the starting index of a slice

“Stop” specifies the ending element of a slice

You can use one of these if you want to skip certain items

image.png

```python
# insert at the start using slicing
x = ['a','b','c','d']
y = [1,2,3]
#x[:0]=y
#print(x)

for element in x:
  y.append(element)
print(y)
```

```python
# insert at the end
x = ['a','b','c','d']
y = [1,2,3]
#y[:0]=x
#print(y)

#using extend function
#x.extend(y)
#print(x)
#Using len()

# for loop
for element in y:
  x.append(element)
print(x)
```

```python
import numpy as np
# print item at first index item
a = np.array([[1,2,3],[4,5,6]])
print(a[0,1])
```

```python
#slicing in string. Print 'Science'
Course = 'Data_Science'
print(Course[5:])
```

```python
#Slicing in list,print item from 2-5 index
lst= [1,2,3,4,5,6]
print(lst[2:5])
```

```python
#Slicing in tuple. Print item from index 2-5
tuple =(1,2,3,4,5,6)
print(tuple[2:5])
```

```python
#Print reverse of all the above items
#String
Course = 'Data_Science'
#print(Course[::-1])
print(Course[-7:-1])
```

```python
#list
lst= [1,2,3,4,5,6]
print(lst[::-1])
```

```python
#tuple
tuple =(1,2,3,4,5,6)
print(tuple[::-1])
```

```python
#skiping alternate items
#string
print(Course[0:12:2])
```

```python
#list
print(lst[0:5:2])
```

```python
#tuple
print(tuple[0:5:2])
```

**Slice function in Python**

```python
#String
print(Course[slice(1,4)])
```

```python
#list
print(lst[slice(1,4)])
```

```python
#tuple
print(tuple[slice(1,4)])
```
